Слово — это абстрактная лингвистическая единица с несколькими определениями
- Ортографическое слово (орфографическое)
  - Последовательность символов между пробелами/знаками препинания
    - Примеры: "кот", "бежал", "по-русски", "что-то"
- Морфологическое слово (лексема)
  - Набор всех словоформ одной лексемы
    - Пример: {бежать, бегу, бежишь, бежал, беги} = одно морфологическое слово
- Синтаксическое слово (словоформа)
  - Конкретная грамматическая форма слова в предложении
    - Пример: "бежал" (прош.вр., муж.р., ед.ч.)
- Фонологическое слово
  - Единица произношения с одним ударением
    - Пример: "по́-русски" может быть одним фонологическим словом

Токен — это элементарная единица обработки текста в компьютерных системах.

Токенизация — это процесс разделения текста на токены согласно определённым правилам, которые могут быть разными.

Ключевое понимание: слово — это лингвистическая концепция о том, как люди организуют язык, а токен — это вычислительная концепция о том, как машины обрабатывают текст. Токенизация — это мост между человеческим языком и машинным пониманием.

```python
def удалить_ненужные_токены_из_предложения(span: Span, tockens_for_del) -> Doc:
    # Фильтруем токены (оставляем только нужные)
    filtered_tokens = [token for token in span if token.text not in tockens_for_del]

    # Создаём новый Doc из отфильтрованных токенов
    # Важно: Doc требует vocab и список строк (или ids), а также spaces (пробелы после токена)
    words = [token.text for token in filtered_tokens]
    spaces = [token.whitespace_ != '' for token in filtered_tokens]

    # Новый Doc (без лишних токенов)
    return nlp_eng(Doc(span.vocab, words=words, spaces=spaces))
```

```python
текущее_предложение = удалить_ненужные_токены_из_предложения(предложения[0], {'#', '\n\n'})
текущее_предложение
```

Начнём с разделения по предлогу `for`. Получается если есть не больше одного `for` (пока сделаем такую версию, но нужно будет доделать для многих `for` в предложении), то делим предложения на 2 части, правую и левую от for — это значит будет 2 блока и 1 связь, которая ведёт от левой части к правой части.

```python
исходное_пр = предложения_после_деленния_по_союзам[0].lower()
целевое_пр = 'the aging mechanisms have been refuted'

целевые_тк = nlp_eng(целевое_пр)
displacy.render(целевые_тк, style="dep", options={'compact': True}, jupyter=True)
```

```python
словарь_самоуказателей_кз = [
    'the purposes of' # цель — в КЗ это мета сущность, обозначающая блок
    'methods of' # метод — в КЗ это мета сущность, обозначающая действия, а значит блок и некоторая цепочка блоков, которую можно назвать метод
]
```

```python
if ' for ' in исходное_пр:
    разделение = исходное_пр.split(' for ')
    левая_часть_предложения: Doc = nlp_eng(разделение[0])
    правая_часть_предложения: Doc = nlp_eng(разделение[1])

print(левая_часть_предложения)
print(правая_часть_предложения)
```

```python
matcher = Matcher(nlp_eng.vocab)
```

```python
pattern = [
    {"LOWER": {"IN": ["methods", "purposes"]}},  # methods или purposes
    {"LOWER": "of"}
]
matcher.add("NOUN_OF", [pattern])
```

```python
matches = matcher(левая_часть_предложения)

for match_id, start, end in matches:
    print("Найдено:", левая_часть_предложения[start:end].text)

spans_to_remove = [левая_часть_предложения[start:end] for match_id, start, end in matches]
# Получаем индексы токенов, которые нужно удалить
remove_token_idxs = set(idx for span in spans_to_remove for idx in range(span.start, span.end))
# Собираем новый текст без этих токенов
filtered_tokens = [token.text_with_ws for i, token in enumerate(левая_часть_предложения) if i not in remove_token_idxs]
filtered_text = nlp_eng("".join(filtered_tokens).strip())
filtered_text
```

"Опровергнут" — это краткая форма страдательного причастия прошедшего времени.

```python
def noun_to_verb(noun):
    for syn in wn.synsets(noun, pos=wn.NOUN):
        if syn is not None:
            for lemma in syn.lemmas():
                if lemma is not None:
                    deriv = lemma.derivationally_related_forms()
                    for d in deriv:
                        if d.synset().pos() == 'v':
                            return d.name()
    return None
```

```python
noun = filtered_text[0].text
verb = noun_to_verb(noun)

if verb:
    print(verb)  # 'refute'
else:
    print(f'Ошибка: слово {noun} не преобразовано в глагол')
```
